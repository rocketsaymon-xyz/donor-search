```python
#изменяем ['first_donation_date'] убираем пропуски
df_donation['first_donation_date'] = (
    df_donation
    .groupby('user_id')['donation_date']
    .transform('min'))

#фильтруем датафрейм. отсекаем годы с менее 10 донациями, как ошибочные
df_donation_filt_year = df_donation[
    df_donation['donation_date'].dt.year.between(1992, 2023)].copy()

#first_donation_date изменим столбец
df_donation_filt_year['first_don_year'] = df_donation_filt_year['first_donation_date'].dt.year

#меняем столбец first_donation_date, отсекаем ошибочные минимальные даты
df_donation_filt_year['first_donation_date'] = (
    df_donation_filt_year.groupby('user_id')['donation_date'].transform('min'))
#считаем сколько удалили данных
filter_count = len(df_donation) - len(df_donation_filt_year)
filter_count_perc = (len(df_donation) - len(df_donation_filt_year)) / len(df_donation)
print(f'количество удаленных строк, которые отфильтрованы в df_donation: {filter_count}')
print(f'процент строк, которые отфильтрованы в df_donation: {filter_count_perc:.2%}')
```
