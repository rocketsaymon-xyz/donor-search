```python
#Функция для корректировки типов данных
def convert_dates(df, date_columns):
    """
    Преобразует указанные столбцы в datetime.
    """
    for col in date_columns:
        df[col] = pd.to_datetime(
            df[col],
            errors='coerce')
    return df
```

```python
#делаем новый датафрейм с необходимыми столбцами
df_activation = df_user_anon_data[['id', 'donor_status']].copy()

#вычисляем пользователей через группировку
df_group_activation = df_activation.groupby('donor_status', as_index = False).agg(
    count_donor = ('id', 'count'))

#считаем доли пользователей
df_group_activation['share'] = (
    df_group_activation['count_donor']
    / df_group_activation['count_donor'].sum()
    * 100)    
```

```python
def plot_stacked_share(
    data,
    share_col,
    count_col,
    label_col,
    title,
    xlabel='Share of users (%)',
    bar_label='Users',
    name_images = 'default'
):

    fig, ax = plt.subplots(figsize=(10, 2))

    left = 0

    for _, row in data.iterrows():

        ax.barh(
            y=bar_label,
            width=row[share_col],
            left=left,
            label=row[label_col]
        )

        ax.text(
            left + row[share_col] / 2,
            0,
            (
                f"{row[share_col]:.1f}%\n"
                f"({row[count_col]:,})"
            ),
            ha='center',
            va='center'
        )

        left += row[share_col]

    ax.set_xlim(0, 100)

    ax.set_title(title)

    ax.set_xlabel(xlabel)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.legend(title=label_col)

    plt.tight_layout()
    
    plt.savefig(
    IMG_DIR / name_images,
    dpi=300,
    bbox_inches='tight')
    
    plt.show()
```

```python
#визуализация
plot_stacked_share(
    data=df_group_activation,
    share_col='share',
    count_col='count_donor',
    label_col='donor_status',
    title='Activation Rate: Registration → First Confirmed Donation',
    xlabel='Share of users (%)',
    name_images = 'Activation_Rate_All.png'
)
```

```python
# Activation Rate by Year (за последние 10 лет)
```

```python
#фильтурем датафрейм по переиоду и отбираем нужные столбцы 
df_activation_10_years = df_user_anon_data.loc[df_user_anon_data['registration_date'].between('2013-11-21' , '2023-11-28')
    , ['id', 'donor_status', 'registration_date']].copy()

#добавляем столбец с когортой по годам
df_activation_10_years['cohort_year'] = df_activation_10_years['registration_date'].dt.year

#вычисляем пользователей через группировку
df_group_activation_10_year = df_activation_10_years.groupby(['cohort_year', 'donor_status'], as_index = False).agg(
    count_donor = ('id', 'count'))

#считаем доли пользователей
df_group_activation_10_year['share_donors'] = (
    df_group_activation_10_year['count_donor']
    / (
    df_group_activation_10_year.groupby('cohort_year')['count_donor'].transform('sum'))
    * 100)    

#фильтруем таблицу по донорами
df_group_activation_10_year = df_group_activation_10_year[df_group_activation_10_year['donor_status'] == 'является донором']
```

```python
### итоговый файл активации по годам
df_group_activation_10_year
```

```python
#делаем визуализацию доноров
fig, ax = plt.subplots(figsize=(12, 6))

mean_share = df_group_activation_10_year['share_donors'].mean()
median_share = df_group_activation_10_year['share_donors'].median()

ax.plot(
    df_group_activation_10_year['cohort_year'],
    df_group_activation_10_year['share_donors'],
    marker='o',
    linewidth=2,
    color='blue',
    label='Activation Rate'
)

# подписи точек
for x, y, n in zip(
    df_group_activation_10_year['cohort_year'],
    df_group_activation_10_year['share_donors'],
    df_group_activation_10_year['count_donor']
):
    ax.text(
        x,
        y + 1,
        f'{y:.1f}%\n(n={n:,})',
        ha='center',
        va='bottom',
        fontsize=9,
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.7)
    )

# тренд
z = np.polyfit(
    df_group_activation_10_year['cohort_year'],
    df_group_activation_10_year['share_donors'],
    1
)

p = np.poly1d(z)

ax.plot(
    df_group_activation_10_year['cohort_year'],
    p(df_group_activation_10_year['cohort_year']),
    linestyle='--',
    linewidth=2,
    color='green',
    alpha=0.7,
    label='Donor Trend'
)

# блок со средним и медианой
ax.text(
    0.02,
    0.98,
    f'Mean: {mean_share:.1f}%\nMedian: {median_share:.1f}%',
    transform=ax.transAxes,
    ha='left',
    va='top',
    fontsize=11,
    bbox=dict(facecolor='white', edgecolor='gray', alpha=1),
    zorder=100
)

ax.set_title('Donor Share by Cohort Year')
ax.set_xlabel('Cohort Year')
ax.set_ylabel('Share of Users (%)')
ax.set_yticks([])

ax.legend()
ax.grid(alpha=0.3)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tight_layout()

plt.savefig(
IMG_DIR / "Activation_Rate_cohort_year.png",
dpi=300,
bbox_inches='tight')

plt.show()
```

```python
#Активация во время проведения ивентов
```

```python
#Объеждиняем датасеты с донорами и ивентами
df_donation_in_event = pd.merge(
    df_activation_not_in_2018[['id', 'donor_status', 'registration_date', 'confirmed_donations']]
    , df_events[['donor_id', 'event_begin', 'event_end']] 
    , how = 'left'
    , left_on= 'id'
    , right_on= 'donor_id').drop(columns='donor_id')

#добавляем столбец, там где начала ивента совпадает с датой регистрацией
reg_date = df_donation_in_event['registration_date'].dt.normalize()
event_begin = df_donation_in_event['event_begin'].dt.normalize()
event_end = df_donation_in_event['event_end'].dt.normalize()

df_donation_in_event['registration_in_event_flag'] = (
    (reg_date >= event_begin)
    & (reg_date <= event_end)
)

# появились дубли. производим маркеровку по булевому столбцу. чтобы выявить тех кто пришел через ивент 
df_donation_in_event['registration_in_event_flag'] = df_donation_in_event.groupby('id')['registration_in_event_flag'].transform('max')

#убираем дубли
df_donation_in_event = df_donation_in_event.drop_duplicates(subset='id')
```

```python
#группируем данные по флагу и считаем user_id
group_donation_in_event = df_donation_in_event.groupby('registration_in_event_flag', as_index=False).agg(
    users_count = ('id', 'count')
    , donors=('confirmed_donations', lambda x: (x > 0).sum()))
#считаем конверсию
group_donation_in_event['conversion_rate'] = group_donation_in_event['donors'] / group_donation_in_event['users_count'] * 100
group_donation_in_event
```

```python
#Строим визуализацию
plot_data = (
    group_donation_in_event
    .set_index('registration_in_event_flag')['donors']
)

plt.figure(figsize=(8, 8))

plt.pie(
    plot_data,
    labels=['Independent Donor', 'Event Donor'],
    autopct='%1.2f%%'
)

plt.title('Доля пользователей по типу регистрации')
plt.show()
```

3. Конверсия участников мероприятий выше на 18,0 процентных пунктов или примерно в 1,5 раза относительно пользователей, зарегистрировавшихся вне мероприятий;

Рекомендации:
1. Разработать стратегию привлечения новых доноров через мероприятия (events), поскольку пользователи, зарегистрировавшиеся в период проведения мероприятий, демонстрируют более высокий Conversion Rate. Основные направления развития:
    - протестировать различные механики мотивации (бонусы, мерч, розыгрыши, социальное признание);
    - внедрить реферальные программы по типу «Приведи друга»;
    - использовать таргетированные рекламные кампании на аудитории, схожие с действующими донорами по полу, возрасту, региону и другим характеристикам;
    - масштабировать наиболее эффективные форматы мероприятий и регулярно оценивать их влияние на Activation Rate, Repeat Rate и Retention.
2. Такой подход позволит увеличить не только количество регистраций, но и долю пользователей, доходящих до первой донации.

## Общие выводы по активации пользователей:
1. Основная проблема проекта находится на этапе «Регистрация → Первая донация». Из всех зарегистрированных пользователей только 23,4% совершили хотя бы одну подтверждённую донацию, тогда как 76,6% пользователей не дошли до этапа активации. Это самый крупный разрыв во всей воронке жизненного цикла донора и главный источник потенциального роста проекта. с 2019 год показатель лучше 35,9%, но 65% пользватаелей не начинают работать с программмой;
2. Эффективность активации существенно улучшилась за последние годы. После исключения эффекта запуска платформы в 2018 году наблюдается устойчивый рост Activation Rate. Доля пользователей, совершивших первую донацию, выросла с 17–19% в ранних когортах до 43–45% в когортах 2022–2023 годов. Это говорит о том, что проект стал лучше привлекать не просто регистрации, а пользователей с реальным намерением стать донорами;
3. Рост происходит не только за счёт количества регистраций, но и за счёт качества привлечения. Наблюдается увеличение как абсолютного количества новых доноров, так и доли пользователей, доходящих до первой донации. Это означает, что улучшения связаны не только с масштабом проекта, но и с более эффективной работой механизмов привлечения и активации. В 22-23 годы приходила в среднем по 7000 новых доноров;
4. На активацию влияют внешние факторы. Анализ месячных когорт показал значительную вариативность Activation Rate — от 30% до 57%. Наиболее высокие показатели наблюдаются весной 2022 и 2023 годов. Это позволяет предположить влияние:
    - сезонности;
    - маркетинговых кампаний;
    - донорских акций;
    - проведение ивентов.
Данный фактор требует отдельного исследования для выявления наиболее эффективных каналов привлечения доноров;
6. Мероприятия привлекают более качественную аудиторию. При более высокой показатели активации
    - Activation Rate участников мероприятий — 53,9% (Разница составляет +18 п.п. или около 1,5 раза.)
    - Activation Rate остальных пользователей — 35,9%
Привлекают крайне мало новых пользователей, а значит работают не эффективно;
7. Основной резерв роста проекта находится среди уже привлечённых пользователей. Проекту не обязательно резко увеличивать объём новых регистраций. В базе уже находится значительное количество пользователей, которые зарегистрировались, но так и не совершили первую донацию.
Даже небольшое улучшение конверсии на этапе активации может привести к существенному росту числа доноров без увеличения затрат на привлечение новой аудитории.
