# Какова доля зарегистрированных пользователей без подтверждённых донаций?

```python
#делаем новый датафрейм с необходимыми столбцами
df_activation = df_user_anon_data[['id', 'donor_status']].copy()

#вычисляем пользователей через группировку
df_group_activation = df_activation.groupby('donor_status', as_index = False).agg(
    count_donor = ('id', 'count'))

#считаем доли пользователей
df_group_activation['share'] = (
    df_group_activation['count_donor']
    / df_group_activation['count_donor'].sum()
    * 100
)    
```

```python
def plot_stacked_share(
    data,
    share_col,
    count_col,
    label_col,
    title,
    xlabel='Share of users (%)',
    bar_label='Users'
):
    """
    Горизонтальная stacked bar диаграмма.

    Parameters
    ----------
    data : DataFrame
    share_col : str
        Столбец с долей (%)
    count_col : str
        Столбец с количеством пользователей
    label_col : str
        Столбец с категориями
    title : str
        Заголовок графика
    xlabel : str
        Подпись оси X
    bar_label : str
        Подпись полосы
    """

    fig, ax = plt.subplots(figsize=(10, 2))

    left = 0

    for _, row in data.iterrows():

        ax.barh(
            y=bar_label,
            width=row[share_col],
            left=left,
            label=row[label_col]
        )

        ax.text(
            left + row[share_col] / 2,
            0,
            (
                f"{row[share_col]:.1f}%\n"
                f"({row[count_col]:,})"
            ),
            ha='center',
            va='center'
        )

        left += row[share_col]

    ax.set_xlim(0, 100)

    ax.set_title(title)

    ax.set_xlabel(xlabel)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.legend(title=label_col)

    plt.tight_layout()
    plt.show()
```

```python
#визуализация
plot_stacked_share(
    data=df_group_activation,
    share_col='share',
    count_col='count_donor',
    label_col='donor_status',
    title='Activation Rate: Registration → First Confirmed Donation',
    xlabel='Share of users (%)'
)
```

```python
#фильтурем датафрейм по переиоду и отбираем нужные столбцы 
df_activation_10_years = df_user_anon_data.loc[df_user_anon_data['registration_date'].between('2013-11-21' , '2023-11-28')
    , ['id', 'donor_status', 'registration_date']].copy()

#добавляем столбец с когортой по годам
df_activation_10_years['cohort_year'] = df_activation_10_years['registration_date'].dt.year

#вычисляем пользователей через группировку
df_group_activation_10_year = df_activation_10_years.groupby(['cohort_year', 'donor_status'], as_index = False).agg(
    count_donor = ('id', 'count'))

#считаем доли пользователей
df_group_activation_10_year['share_donors'] = (
    df_group_activation_10_year['count_donor']
    / (
    df_group_activation_10_year.groupby('cohort_year')['count_donor'].transform('sum'))
    * 100)    

#фильтруем таблицу по донорами
df_group_activation_10_year = df_group_activation_10_year[df_group_activation_10_year['donor_status'] == 'является донором']
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

mean_share = df_group_activation_10_year['share_donors'].mean()
median_share = df_group_activation_10_year['share_donors'].median()

ax.plot(
    df_group_activation_10_year['cohort_year'],
    df_group_activation_10_year['share_donors'],
    marker='o',
    linewidth=2,
    color='blue',
    label='Activation Rate'
)

# подписи точек
for x, y, n in zip(
    df_group_activation_10_year['cohort_year'],
    df_group_activation_10_year['share_donors'],
    df_group_activation_10_year['count_donor']
):
    ax.text(
        x,
        y + 1,
        f'{y:.1f}%\n(n={n:,})',
        ha='center',
        va='bottom',
        fontsize=9,
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.7)
    )

# тренд
z = np.polyfit(
    df_group_activation_10_year['cohort_year'],
    df_group_activation_10_year['share_donors'],
    1
)

p = np.poly1d(z)

ax.plot(
    df_group_activation_10_year['cohort_year'],
    p(df_group_activation_10_year['cohort_year']),
    linestyle='--',
    linewidth=2,
    color='green',
    alpha=0.7,
    label='Donor Trend'
)

# блок со средним и медианой
ax.text(
    0.02,
    0.98,
    f'Mean: {mean_share:.1f}%\nMedian: {median_share:.1f}%',
    transform=ax.transAxes,
    ha='left',
    va='top',
    fontsize=11,
    bbox=dict(facecolor='white', edgecolor='gray', alpha=1),
    zorder=100
)

ax.set_title('Donor Share by Cohort Year')
ax.set_xlabel('Cohort Year')
ax.set_ylabel('Share of Users (%)')
ax.set_yticks([])

ax.legend()
ax.grid(alpha=0.3)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
#функция для гистограммы
def plot_conversion(
    data,
    conversion_col,
    users_col,
    labels,
    title,
    ylabel='Conversion Rate (%)'
):
    """
    Строит conversion chart + uplift + размер групп + odds ratio

    data — датафрейм после группировки
    conversion_col — столбец с conversion (%)
    users_col — размер группы
    labels — подписи столбцов
    title — заголовок
    """

    values = data[conversion_col].values
    users = data[users_col].values
    fig, ax = plt.subplots(figsize=(9, 5))
    bars = ax.bar(labels, values)

    # подписи процентов и размера группы
    for bar, val, n in zip(bars, values, users):
        ax.text(
            bar.get_x() + bar.get_width()/2,
            val + 1,
            (f'{val:.1f}%\n'
                f'(n={n:,})'), ha='center')

     # uplift и ratio
    uplift = (values[1] - values[0]) / values[0] * 100 
    ratio = values[1] / values[0]

    # odds ratio
    p0 = values[0] / 100
    p1 = values[1] / 100
    odds_ratio = (p1 / (1 - p1)) / (p0 / (1 - p0))

    # блок справа сверху
    ax.text(0.98, 0.95,
        (f'Uplift: +{uplift:.0f}%\n'
            f'Ratio: ×{ratio:.1f}\n'
            f'Odds Ratio: {odds_ratio:.2f}'),
        transform=ax.transAxes,
        ha='right',
        va='top',
        fontsize=11,
        bbox=dict(
            facecolor='white',
            alpha=0.8))

    # оформление
    ax.set_yticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # подпись оси без значений
    ax.set_ylabel(ylabel)
    ax.set_ylim(0, max(values) * 1.25)
    ax.set_title(title)
    plt.tight_layout()
    plt.show()
```

```python
#строим гистограмму конверсии 
labels = ['No blood info','Blood specified']
title = 'First Donation Conversion by Blood Information Availability'
plot_conversion(group_cr_blood, 'conversion_rate', 'users_count', labels, title)
```

# Как ведут себя доноры после первой донации? Определение Churn Threshold.


```python
#считаем период между первой и второй донацией
```

```python
#фильтруем датафрейм, чтобы оставить только 2019 для исследования (чтобы не было пандемии и СВО)
df_donation_churn_2019 = df_donation_filt_year[df_donation_filt_year['first_donation_date'].between('2019-01-01', '2019-12-31')].copy()

#считаем время донации от первой
df_donation_churn_2019['period_day_donation'] = (df_donation_churn_2019['donation_date'] -
                                                 df_donation_churn_2019['first_donation_date']).dt.days
```

```python
#оставляем только 1 и 2 донацию
df_donation_churn_2019_2 = df_donation_churn_2019[(df_donation_churn_2019['number_donation'] <=2)]

#через pivot разварачиваем данные
pivot_donation_churn_2 = (df_donation_churn_2019_2.pivot(
        index='user_id',
        columns='number_donation',
        values='donation_date').reset_index())

#переименовываем столбцы
pivot_donation_churn_2.columns = ['user_id', 'first_donation', 'second_donation']

# считаем разницу между донациям
pivot_donation_churn_2['days_to_second'] = (pivot_donation_churn_2['second_donation'] - 
                                          pivot_donation_churn_2['first_donation']).dt.days
```

```python
#считаем days_per_returned_pct (сколько требуется дней для роста на каждый 1%)
#задаем квантили
quantiles = [i / 100 for i in range(0, 101, 5)]

# убираем user_id с 1 донацией, добавляем квантили
quantiles_donation_churn = (pivot_donation_churn_2['days_to_second'].dropna()
    .quantile(quantiles)
    .reset_index())

#переименовываем столбцы
quantiles_donation_churn.columns = ['returned_share', 'days']

# считаем прирост returned_share
quantiles_donation_churn['delta_users_pct'] = (
    quantiles_donation_churn['returned_share'].diff())

# считаем прирост days
quantiles_donation_churn['delta_days'] = (
    quantiles_donation_churn['days'].diff())

# сколько дней нужно ждать для следующего 1% пользователя
quantiles_donation_churn['days_per_returned_pct'] = (quantiles_donation_churn['delta_days'] /
    (quantiles_donation_churn['delta_users_pct'] * 100))
```

```python
# Cumulative Return to Second Donation строим линейый график


fig, ax = plt.subplots(figsize=(12, 6))

# данные
x = quantiles_donation_churn['days']
y = quantiles_donation_churn['returned_share'] * 100

# основная линия
ax.plot(x, 
        y,
        marker='o', 
        linewidth=2)

# вертикальные линии от точек к оси X
ax.vlines(
    x=x,
    ymin=0,
    ymax=y,
    linewidth=1,
    alpha=0.3)

# подписи на линии с шагом 20%
labels = [20, 40, 60, 80]

for level in labels:
    idx = (y - level).abs().idxmin()
    ax.text(
        x.loc[idx],
        y.loc[idx] + 3,
        f'{level}%\n{x.loc[idx]:.0f} d',
        ha='center',
        va='bottom',
        bbox=dict(
            boxstyle='round,pad=0.4',
            facecolor='#DFF5E1',
            edgecolor='#6DAA75',
            linewidth=1))

# линия выбранного churn threshold
ax.axhline(
    y=80,
    linestyle='--',
    linewidth=1)

ax.axvline(
    x=210,
    linestyle='--',
    linewidth=1)

ax.text(
    215,
    75,
    'Churn threshold ≈ 210 d',
    va='center')

# оформление
ax.set_title('Cumulative Return to Second Donation')

ax.set_xlabel('Days after first donation')

# название оси Y оставляем, шкалу убираем
ax.set_ylabel('Returned users (%)')
ax.set_yticks([])

# шкала X с шагом 100 дней
ax.set_xticks(
    np.arange(
        0,
        x.max() + 100,
        100))

# убрать лишние рамки
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
#Определяем период для Churn Threshold
```

```python
#задаем последнюю дату отчета
analysis_date = pd.Timestamp('2023-11-28')

#фильтруем датафрейм, чтобы убрать донации 486 дней (50% донаций уже сделали по Donation Frquency)
cutoff_date = analysis_date - pd.Timedelta(days=486)

df_churn_period = (df_donation_filt_year[
        df_donation_filt_year['donation_date'] <= cutoff_date
    ]
    .copy())

df_churn_period = (df_churn_period[
        df_churn_period['days_between'] != 0
    ])

#Задаем флаг донации
df_churn_period['factor_donation'] = (df_churn_period['days_between'].notna()
    .astype(int))

#задаем период до даты отчета
df_churn_period['duration'] = np.where(
    df_churn_period['days_between'].notna(),
    df_churn_period['days_between'],
    (analysis_date
        - df_churn_period['donation_date']).dt.days)
```

```python
kmf = KaplanMeierFitter()

kmf.fit(
    durations=df_churn_period['duration'],
    event_observed=df_churn_period['factor_donation']
)
```

```python
survival = kmf.survival_function_.reset_index()
survival.columns = ['days', 'survival']

# ограничиваем область анализа, чтобы убрать возвраты через 5 лет
min_days = 7 #меньше этой даты анамольные донации
max_days = 2000 #свыше не будем рассматривать

survival_knee = survival[(survival['days'] > min_days) &
 (survival['days'] <= max_days)].copy()

kneedle = KneeLocator(
    x=survival_knee['days'],
    y=survival_knee['survival'],
    curve='convex',
    direction='decreasing'
)

hard_churn_day = int(kneedle.knee)


survival_at_churn = (
    survival_knee.loc[
        survival_knee['days'] <= hard_churn_day,
        'survival'
    ]
    .iloc[-1]
)

print(
    f'Hard Churn: {hard_churn_day} дней\n'
    f'Survival: {survival_at_churn*100:.1f}%'
)
```

```python
# Таблица survival
survival = kmf.survival_function_.reset_index()
survival.columns = ['days', 'survival']

# Ограничиваем область анализа
survival_3000 = survival[
    survival['days'] <= 3000
].copy()

# Survival thresholds
s20 = survival.loc[
    survival['survival'] <= 0.20,
    'days'
].min()

s30 = survival.loc[
    survival['survival'] <= 0.30,
    'days'
].min()

# Hard Churn point
survival_at_churn = (
    survival.loc[
        survival['days'] <= hard_churn_day,
        'survival'
    ]
    .iloc[-1]
)

print(f'Survival 20%: {s20:.0f} дней')
print(f'Survival 30%: {s30:.0f} дней')
print(f'Hard Churn: {hard_churn_day:.0f} дней')
print(f'Survival: {survival_at_churn*100:.1f}%')
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

# Kaplan-Meier
ax.plot(
    survival_3000['days'],
    survival_3000['survival'] * 100,
    linewidth=3,
    label='Kaplan-Meier'
)

# Survival levels
for day, level, color in [
    (s20, 20, 'orange'),
    (s30, 30, 'green')
]:

    ax.axhline(
        y=level,
        linestyle='--',
        alpha=0.6,
        color=color
    )

    ax.axvline(
        x=day,
        linestyle='--',
        alpha=0.6,
        color=color
    )

    ax.text(
        day + 20,
        level + 2,
        f'{level}% = {day:.0f} дн.',
        color=color,
        fontsize=11,
        fontweight='bold'
    )

# Hard Churn
ax.axvline(
    hard_churn_day,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Hard Churn = {hard_churn_day:.0f} дн.'
)

ax.scatter(
    hard_churn_day,
    survival_at_churn * 100,
    s=120,
    color='red',
    zorder=5
)

ax.annotate(
    (
        f'Hard Churn\n'
        f'{hard_churn_day:.0f} дн.\n'
        f'{survival_at_churn*100:.1f}%'
    ),
    xy=(hard_churn_day, survival_at_churn * 100),
    xytext=(
        hard_churn_day + 100,
        survival_at_churn * 100 + 10
    ),
    arrowprops=dict(arrowstyle='->'),
    fontsize=11,
    bbox=dict(boxstyle='round,pad=0.4')
)

ax.set_xlim(0, 3000)

ax.set_title(
    'Kaplan-Meier Survival Curve\n(Return to Next Donation)',
    fontsize=16,
    fontweight='bold'
)

ax.set_xlabel('Дни после предыдущей донации')
ax.set_ylabel('Вероятность отсутствия следующей донации (%)')

ax.grid(alpha=0.3)
ax.legend()

plt.tight_layout()
plt.show()
```

```python

#выбираем период с 2010 по 2022 год, по годовым когортам
df_retention_year = df_donation_filt_year[df_donation_filt_year['first_don_year'].between(2010, 2022)].copy()
```

```python
# добавляем столбец когорту с месяцем и годом донации
df_retention_year['cohort_month'] = (df_retention_year['donation_date'].dt.to_period('M'))

#считаем дни между первой донацией и другими
df_retention_year['period_day'] = (df_retention_year['donation_date'] - df_retention_year['first_donation_date']).dt.days

df_retention_year.head(2)
```

```python
# 1. Находим максимальный period_day для каждого пользователя (последняя активность)
user_last_activity = (df_retention_year.groupby('user_id')
                      .agg(last_period_day=('period_day', 'max'))
                      .reset_index())

# 2. Присоединяем
df_retention_year = df_retention_year.merge(user_last_activity, on='user_id', how='left')

# 3. Функция для Rolling Retention
def calculate_rolling_retention(df, horizons=[191, 245, 500]):
    result = []
    
    for cohort_year in sorted(df['first_don_year'].unique()):
        cohort = df_retention_year[df_retention_year['first_don_year'] == cohort_year].copy()
        total_users = cohort['user_id'].nunique()
        
        row = {'first_don_year': cohort_year, 'cohort_size': total_users}
        
        for horizon in horizons:
            active_users = cohort[cohort['last_period_day'] >= horizon]['user_id'].nunique()
            rolling_ret = round(active_users / total_users * 100, 2) if total_users > 0 else 0
            row[f'RR_rolling_{horizon}d'] = rolling_ret
        
        result.append(row)
    
    return pd.DataFrame(result)


# Использование
rr_rolling = calculate_rolling_retention(df_retention_year, horizons=[191, 245, 500])
```

```python
#
df_user_bonus_2019 = df_user_anon_data[['id', 'confirmed_donations', 'count_bonuses_taken']].copy()

#создаем новый столбец,  в котором будет доля полученных бонусов от количества донаций
# Условия для проверки
conditions = [
    (df_user_bonus_2019['confirmed_donations'] == 0) & (df_user_bonus_2019['count_bonuses_taken'] > 0),
    (df_user_bonus_2019['confirmed_donations'] == 0) & (df_user_bonus_2019['count_bonuses_taken'] == 0)
]

# Значения, которые присваиваются при выполнении условий
choices = [1001,  # 101% если донатов 0, а бонусы есть
    0.00]   # 0% если и донатов 0, и бонусов 0 (чтобы избежать ошибок)


# Если ни одно условие не выполнено, срабатывает стандартная формула (default)
df_user_bonus_2019['fraud_bonus'] = np.select(
    conditions, 
    choices, 
    default=(df_user_bonus_2019['count_bonuses_taken'] / df_user_bonus_2019['confirmed_donations'] * 100))

#Соритровка датафрейма по столбцу
df_user_bonus_2019.sort_values(by='fraud_bonus', inplace=True)

#создаем фрод маркер
df_user_bonus_2019['fraud_marker'] = np.where(
    df_user_bonus_2019['fraud_bonus'] >= 100, 1, 0)

#группируем данные по маркеру
group_fruad_donation = df_user_bonus_2019.groupby(['fraud_marker'], as_index=False).agg(
    count_id = ('id', 'count')
    , sum_confirmed_donations = ('confirmed_donations', 'sum')
    , sum_bonuses_taken = ('count_bonuses_taken', 'sum')).assign(
        donations_per_user=lambda x:
            x['sum_bonuses_taken'] / x['sum_confirmed_donations'])
group_fruad_donation        
```

    Activation Rate вырос с 17–19% в ранних когортах до 43–45% в 2022–2023 годах. Проект стал привлекать более мотивированных пользователей и эффективнее доводить их до первой донации.

3. Полнота профиля тесно связана с вероятностью первой донации;

    Около половины пользователей не возвращаются после первой донации. Именно этот этап формирует основные потери жизненного цикла донора.

    Алгоритм Knee Detection определил, что точка выхода кривой Kaplan-Meier на плато находится:  
Hard Churn Threshold = 245 дней.  
К этому моменту основная часть повторных донаций уже совершена, а скорость появления новых возвратов существенно снижается.

6. Для доноров характерен длинный цикл возврата;
    80% повторных донаций происходят в течение 210–220 дней после первой донации. Для проекта более корректно использовать Rolling Retention (215 дней, который имеет небольшое снижении на протяжении 2 лет активности) и Churn Threshold (245 дней), чем классический Retention Rate с короткими окнами.
