```python
#корень загрузки
BASE_DIR = Path.cwd()
# папки для загрузки
DATA_RAW = BASE_DIR / 'файлы_итоги'
#папки для сохранения
DATA_PROCESSED = BASE_DIR / "python_итог"
#Графики
IMG_DIR = BASE_DIR / 'images'
```

```python
#Загружаем данные
df_events = pd.read_csv(DATA_RAW /'donor_ad_hoc_eda_events_donation.csv', sep=";")
df_donation_filt_year = pd.read_csv(DATA_RAW / 'donor_ad_hoc_eda_donation_anon.csv', sep=";")
df_user_anon_data = pd.read_csv(DATA_RAW / 'donor_ad_hoc_eda_user_anon_data.csv', sep=";")
```

```python
#Функция для корректировки типов данных
def convert_dates(df, date_columns):
    """
    Преобразует указанные столбцы в datetime.
    """
    for col in date_columns:
        df[col] = pd.to_datetime(
            df[col],
            errors='coerce')
    return df
```

```python
#корректируем типы данных
df_user_anon_data = convert_dates(
    df_user_anon_data,
    ['birth_date', 'last_activity', 'registration_date'])

df_donation_filt_year = convert_dates(
    df_donation_filt_year,
    ['donation_date', 'first_donation_date'])

df_events = convert_dates(
    df_events,
    ['event_begin', 'event_end', 'donation_data'])
```

```python
# Какие характеристики отличают пользователей без первой донации?
```

```python
#создаем новый столбец, который будет означать указал ли пользователь группу крови.
df_user_anon_data['blood_flag'] = (~df_user_anon_data['blood_type'].isin(['нет_данных', 'Не знаю'])).astype(int)
df_user_anon_data['birth_date_flag'] = (~df_user_anon_data['birth_date'].isna()).astype(int)
```

```python
#группируем данные по donor_status
df_donor_status_group = df_user_anon_data.groupby('donor_status', as_index=False).agg(
    age_categ = ('age_donor_categ', 'count')
    , birth_date = ('birth_date', 'count')
    , email_phone_donor = ('email_phone_donor', 'sum')
    , networks_vk_google = ('networks_vk_google', 'sum')
    , blood_flag = ('blood_flag', 'sum'))
```

```python
#считаем доли в группах
df_donor_status = pd.DataFrame()
df_donor_status['donor_status'] = df_donor_status_group['donor_status']
df_donor_status['age_categ'] = df_donor_status_group['age_categ']
df_donor_status['birth_date_perc'] = df_donor_status_group['birth_date'] / df_donor_status_group['age_categ'] * 100
df_donor_status['email_phone_perc'] = df_donor_status_group['email_phone_donor'] / df_donor_status_group['age_categ'] * 100
df_donor_status['networks_vk_google_perc'] = df_donor_status_group['networks_vk_google'] / df_donor_status_group['age_categ'] * 100
df_donor_status['blood_flag_perc'] = df_donor_status_group['blood_flag'] / df_donor_status_group['age_categ'] *100
```

```python
#итоговый файл зависимости от активации
df_donor_status.head(5)
```

```python
# === Feature Uplift: Impact of Profile Completeness on First Donation ===
df_plot = df_donor_status.copy()
df_plot = df_plot.set_index('donor_status')

cols = [
    'birth_date_perc',
    'email_phone_perc',
    'networks_vk_google_perc',
    'blood_flag_perc'
]

feature_names = {
    'birth_date_perc': 'birth_date',
    'email_phone_perc': 'contacts_email_phone',
    'networks_vk_google_perc': 'networks_vk_google',
    'blood_flag_perc': 'blood_Profile'
}

donor = df_plot.loc['является донором', cols]
non_donor = df_plot.loc['не_является_донором', cols]

absolute_diff_pp = donor - non_donor
relative_uplift_pct = (absolute_diff_pp / non_donor) * 100

uplift_df = absolute_diff_pp.rename('absolute_diff_pp').to_frame()
uplift_df['relative_uplift_pct'] = relative_uplift_pct
uplift_df = uplift_df.rename(index=feature_names)

# Сортировка по убыванию
uplift_df = uplift_df.sort_values('absolute_diff_pp', ascending=False)

# === График ===
fig, ax = plt.subplots(figsize=(14, 7))   # ← Увеличил размер

bars = ax.barh(
    uplift_df.index,
    uplift_df['absolute_diff_pp'],
    color='#1f77b4'
)

ax.axvline(x=0, color='gray', linestyle='--', linewidth=1)

# Подписи
for i, (abs_value, rel_value) in enumerate(
    zip(uplift_df['absolute_diff_pp'], uplift_df['relative_uplift_pct'])
):
    # pts — в центре столбца
    ax.text(
        abs_value / 2,
        i,
        f'{abs_value:+.1f} pts',
        va='center',
        ha='center',
        fontsize=11,
        fontweight='bold',
        color='white' if abs_value > 8 else 'black'
    )
    
    # (%) — справа от столбца с умным отступом
    offset = 0.5 if abs_value > 0 else 0
    ax.text(
        abs_value + offset,
        i,
        f'({rel_value:+.1f}%)',
        va='center',
        ha='left' if abs_value > 0 else 'right',
        fontsize=10,
        color='darkblue'
    )

# Оформление
ax.set_title('Feature Uplift: Impact of Profile Completeness on First Donation (All Time)', 
             fontsize=15, pad=25)
ax.set_xlabel('Difference in Share of Users with Filled Feature (Donors - Non-Donors), % points', 
              fontsize=12)

# Информация о выборке
total_users = int(df_donor_status['age_categ'].sum()) if 'age_categ' in df_donor_status.columns else 'N/A'
ax.text(0.02, -0.15, f'Total users: {total_users:,.0f} | Period: All Time', 
        transform=ax.transAxes, fontsize=11, style='italic')

x_min = uplift_df['absolute_diff_pp'].min()
x_max = uplift_df['absolute_diff_pp'].max()

ax.set_xlim(
    x_min - 5,
    x_max + 10
)

plt.subplots_adjust(left=0.30, right=0.95, bottom=0.25)

plt.savefig(
IMG_DIR / "Feature_Uplift_Impact_of_Profile_Completeness_on_First_Donation_(All_Time).png",
dpi=300,
bbox_inches='tight')


plt.show()
```

## выводы Conversion Uplift donors vs non-donors:

1) Среди пользователей, совершивших хотя бы одну подтверждённую донацию, доля заполненной информации о группе крови оказалась в 3,3 раза выше, чем среди пользователей без подтверждённых донаций. Uplift - 36 пунктов;
2) Доноры в 1,9 раза чаще указывают дату рождения, чем пользователи без подтверждённых донаций.  Uplift - 20,0 пунктов;
3) Доноры в 1,3 раза чаще оставляют контактные данные, чем пользователи без подтверждённых донаций. Uplift - 21,2 пункт;
4) Доля пользователей с привязанными аккаунтами ВКонтакте и Google отличается незначительно между группами, поэтому на текущем этапе не наблюдается выраженной связи между социальной авторизацией и совершением первой донации.
   
* посчитан изначально uplift, чтобы можно было выявить всплески, и строить Retention по чувствительным параметрам.

```python
#выделяем группу по сдачи крови:
group_cr_blood = df_user_anon_data.groupby('blood_flag', as_index=False). agg(
    users_count = ('id', 'count')
    ,  donors=('confirmed_donations', lambda x:
            (x > 0).sum()))
#считаем конверсию
group_cr_blood['conversion_rate'] = (group_cr_blood['donors'] /
    group_cr_blood['users_count']) * 100
group_cr_blood
```

```python
#функция для гистограммы
def plot_conversion(
    data,
    conversion_col,
    users_col,
    labels,
    title,
    ylabel='Conversion Rate (%)'
):
    values = data[conversion_col].values
    users = data[users_col].values
    fig, ax = plt.subplots(figsize=(9, 5))
    bars = ax.bar(labels, values)

    # подписи процентов и размера группы
    for bar, val, n in zip(bars, values, users):
        ax.text(
            bar.get_x() + bar.get_width()/2,
            val + 1,
            (f'{val:.1f}%\n'
                f'(n={n:,})'), ha='center')

     # uplift и ratio
    uplift = (values[1] - values[0]) / values[0] * 100 
    ratio = values[1] / values[0]

    # odds ratio
    p0 = values[0] / 100
    p1 = values[1] / 100
    odds_ratio = (p1 / (1 - p1)) / (p0 / (1 - p0))

    # блок справа сверху
    ax.text(0.98, 0.95,
        (f'Uplift: +{uplift:.0f}%\n'
            f'Ratio: ×{ratio:.1f}\n'
            f'Odds Ratio: {odds_ratio:.2f}'),
        transform=ax.transAxes,
        ha='right',
        va='top',
        fontsize=11,
        bbox=dict(
            facecolor='white',
            alpha=0.8))

    # оформление
    ax.set_yticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # подпись оси без значений
    ax.set_ylabel(ylabel)
    ax.set_ylim(0, max(values) * 1.25)
    ax.set_title(title)
    plt.tight_layout()
    plt.show()
```

```python
#строим гистограмму конверсии 
labels = ['No blood info','Blood specified']
title = 'First Donation Conversion by Blood Information Availability'
plot_conversion(group_cr_blood, 'conversion_rate', 'users_count', labels, title)
```

```python
#Какая первая донация встречается чаще: платная или безвозмездная?
```

```python
#оставляем только необходимые столбцы, плюс делаем сортировку внутрю user_id по датам донации
df_donation_pay = df_donation_filt_year[['user_id', 'donation_date', 'donation_type', 'number_donation']].sort_values(by=['user_id'
                                                                                                       , 'donation_date']).copy()
```

```python
#оставляем только первую донацию
df_donation_pay_first = df_donation_pay[df_donation_pay['number_donation'] == 1]
# группируем датафрейм по donation_type
group_donation_type_first = df_donation_pay_first.groupby('donation_type', as_index=False).agg(
    count_user_id = ('user_id', 'count'))
#считаем доли
group_donation_type_first['share_pay_donation'] =  (
    group_donation_type_first['count_user_id'] /
    group_donation_type_first['count_user_id'].sum() * 100
)
group_donation_type_first
```

```python
# строим график
# данные
labels = group_donation_type_first['donation_type']
values = group_donation_type_first['share_pay_donation']
counts = group_donation_type_first['count_user_id']

# ratio
ratio = round(
    values.iloc[0]
    /
    values.iloc[1],
    1
)

fig, ax = plt.subplots(
    figsize=(9, 5)
)

bars = ax.barh(
    labels,
    values
)

# подписи
for bar, val, cnt in zip(
    bars,
    values,
    counts
):

    ax.text(
        val + 1,

        bar.get_y()
        + bar.get_height()/2,

        f'{val:.1f}%\n(n={cnt:,})',

        va='center'
    )

# ratio-box
ax.text(
    0.98,
    0.95,

    f'Donation Type Ratio\n{ratio}x',

    transform=ax.transAxes,

    ha='right',
    va='top',

    bbox=dict(
        boxstyle='round,pad=0.5',
        facecolor='#E8F5E9',
        edgecolor='#66BB6A'
    )
)

# оформление
ax.set_title(
    'First Donation Type Distribution'
)

# убрать шкалу
ax.set_xticks([])

# убрать лишние рамки
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()

plt.show()
```

## Donation pay_type Ratio Выводы:
1. Безвозмездные донации составляют 90,8% всех первых донаций пользователей.
2. Платные донации составляют только 9,2%.
3. Donation Type Ratio составил 9.9x:
    * безвозмездная первая донация встречается почти в 10 раз чаще, чем платная.
4. Полученные результаты могут свидетельствовать о том, что для большинства пользователей мотивацией первой донации являются:

    - отличаются ли такие пользователи по retention;

    - как меняется вероятность повторной донации в зависимости от типа первой донации.
